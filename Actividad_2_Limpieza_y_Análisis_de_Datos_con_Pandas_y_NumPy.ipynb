```python
#Manejo de valores nulos

from os import pardir

df ['precio'] = df ['precio'].replace('',None)
df ['cantidad'] = df ['cantidad'].replace(' ',None)

df["precio"] = df["precio"].fillna ("sin_informacion")
df["cantidad"] = df["cantidad"].fillna ("sin_informacion")

df.reset_index(drop=True)
```

```python
#Estandarizacion de columnas

#País

df["pais"] = df["pais"].replace(
{"COL" :"Colombia",
 "col":"Colombia",
 "peru":"Perú",
 "chile":"Chile",
})

#metodo de pago

df ["metodo_pago"] = df ["metodo_pago"].replace(
   {"TRANSFERENCIA" :"transferencia",
   })
```

```python
df ["precio"] = pd.to_numeric(df["precio"], errors="coerce")
df ["cantidad"] = pd.to_numeric(df["cantidad"], errors="coerce")
```

```python
#Valores atipicos

df = df [df["precio"]<10000]
```

```python
df.reset_index(drop=True)
```

Parte 3 — Análisis con Pandas
Actividades
• Crear una nueva columna llamada total: df['total'] = df['precio'] * df['cantidad']
Obtener:
• Total vendido
• Promedio de ventas
• Venta máxima
• Venta mínima

```python
#Nueva columna

df["total"] = df["precio"] * df["cantidad"]
df.reset_index(drop=True)
```

```python

print("Total vendido: ", f"${df['total'].sum():,.0f}".replace(",", "."))
print("Promedio de ventas: ", f"${df['total'].mean():,.0f}".replace(",", "."))
print("Venta máxima: ", f"${df['total'].max():,.0f}".replace(",", "."))
print("Venta mínima: ", f"${df['total'].min():,.0f}".replace(",", "."))
```

Parte 4 — NumPy
Actividades
• Convertir columnas numéricas a un array de NumPy: data = df[['precio',
'cantidad']].to_numpy()
• Separar columnas utilizando indexación: precios = data[:, 0]
cantidades = data[:, 1]
• Aplicar vectorización para calcular ventas totales: totales = precios *
cantidades

```python
import numpy as np
```

```python
#Elimina de la tabla el registro que no tiene cantidad

df = df.dropna(subset= ["cantidad"])
df.reset_index(drop=True)
```

```python
data = df[["precio", "cantidad"]].to_numpy()
print(data)
```

```python
precio = data[:,0]
cantidad = data[:,1]
print(precio)
print(cantidad)
```

```python
totales = precio * cantidad
print(totales)
```

Parte 5 — Análisis con NumPy

Utilizando funciones de NumPy, calcular:
• Suma total de ventas
• Promedio de ventas
• Venta máxima
• Cantidad de ventas superiores a 1000

```python

total_ventas = np.sum(totales)
print("Total de ventas:", f"${total_ventas:,.0f}".replace(",", "."))


promedio_ventas = np.mean(totales)
print("Promedio ventas:", f"{promedio_ventas:,.0f}".replace(",", "."))


venta_max = np.max(totales)
print("Venta máxima:", f"${venta_max:,.0f}".replace(",", "."))

cantidad_ventas = np.sum(totales > 1000)
print("Cantidad ventas > 1000:", f"{cantidad_ventas:,.0f}".replace(",","."))
```
