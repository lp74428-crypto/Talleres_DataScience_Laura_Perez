
Tienes una lista de clientes con datos repetidos. Además, cuentas con dos listas: clientes activos e inactivos.

Objetivos:

Eliminar duplicados de la lista original.
Identificar clientes en ambos estados.
Encontrar clientes solo activos.
Recorrer los clientes y clasificarlos.

Pistas:

Usa set() para eliminar duplicados.
Usa intersection() para encontrar coincidencias.
Usa difference() para encontrar diferencias.
Usa un for con if para clasificar.

```python
#Creación de listas

Clientes = ["Marcela", "Carmen", "Kevin", "Franco", "Silvana","Carlos", "Kevin", "Marcela", "Pedro", "Juan"]
Clientes_activos = ["Marcela", "Carmen", "Kevin", "Franco", "Silvana", "Kevin"]
Clientes_inactivos = ["Carlos", "Pedro", "juan", "Silvana", "Marcela",]

#Eliminar duplicados
Clientes_unicos = set(Clientes)
print(Clientes_unicos)
```

```python
#En ambos estados

Clientes_activos_e_inactivos = set(Clientes_activos).intersection(set(Clientes_inactivos))
print(Clientes_activos_e_inactivos)
```

```python
#Encontrar diferencias

Clientes_diferencias =set(Clientes_activos).difference(set(Clientes_inactivos))
print(Clientes_diferencias)
```

```python
#Clasificación con FOR e IF

print("Estado de los clientes <3")

for cliente in Clientes:
    if cliente in Clientes_activos:
        print(f"{cliente} es activo")

```

Ejercicio 3: Estudiantes y Cursos
Contexto:
Tienes estudiantes inscritos en dos cursos distintos.
Objetivos:
• Encontrar estudiantes en ambos cursos.
• Identificar estudiantes exclusivos por curso.
• Unir todos los estudiantes.
• Clasificar estudiantes según su participación.
Pistas:
• Usa & para intersección.
• Usa difference() para exclusivos.
• Usa union() para todos.
• Usa for + if para clasificar.

```python
#Estudiantes y cursos

estudiantes_ingles = {"Laura", "Roberto","Arquimedes", "Jorge"}
estudiantes_español = {"Juan", "Laura", "Valentina", "Arquimedes"}


```

```python
#Estudiantes en ambos cursos

estudiantes_ambos_cursos = estudiantes_ingles & estudiantes_español
print(estudiantes_ambos_cursos)
```

```python
#Estudiantes exclusivo por curso

print("Estudiantes exclusivos por curso --- INGLES ---")

estudiantes_exclusivos_ingles = set(estudiantes_ingles).difference(set(estudiantes_español))
print(estudiantes_exclusivos_ingles)

print("Estudiantes exclusivos por curso --- ESPAÑOL ---")

estudiantes_exclusivos_español = set(estudiantes_español).difference(set(estudiantes_ingles))
print(estudiantes_exclusivos_español)
```

```python
#tODOS LOS ESTUDIANTES

todos_los_estudiantes = set(estudiantes_ingles).union(set(estudiantes_español))
print(todos_los_estudiantes)
```

```python
#Clasificación de estudiantes

for estudiante in todos_los_estudiantes:
    if estudiante in estudiantes_ambos_cursos:
        print(f"{estudiante} estudiante de ingles y español. ")
    elif estudiante in estudiantes_exclusivos_ingles:
        print(f"{estudiante} estudiante de ingles.")
    elif estudiante in estudiantes_exclusivos_español:
        print(f"{estudiante} estudiante de español.")

```
